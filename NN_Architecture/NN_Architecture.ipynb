```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf

from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import resnet50
```

# NN Architecture

```python
backbone = resnet50.ResNet50(include_top = False) # exclude the last layer
```

```python
backbone.summary()
```

```python
model = Sequential([
    backbone,
    Dense(256, activation = "relu"),
    Dropout(0.5),
    Dense(128, activation = "relu"),
    Dense(20, activation = "softmax")
])
```

```python
model
```

```python
model.summary()
```

```python
model_plus = Sequential([
    Input((299, 299, 3)), 
    backbone,
    GlobalAveragePooling2D(),
    Dense(256, activation = "relu"),
    Dropout(0.5),
    Dense(128, activation = "relu"),
    Dense(20, activation = "softmax")
])
```

```python
model_plus.summary()
```

```python
backbone.trainable = False # this prevents backpropagation i.e. updating the weights
```

```python
model_new = Sequential([
    Input((299, 299, 3)), 
    backbone,
    GlobalAveragePooling2D(),
    Dense(256, activation = "relu"),
    Dropout(0.5),
    Dense(128, activation = "relu"),
    Dense(20, activation = "softmax")
])
```

```python
model_new.summary()
```

```python
backbone.trainable = True
```

```python
backbone.layers[1:20]
```

```python
for layer in backbone.layers[1:60]:
    layer.trainable = False
```
